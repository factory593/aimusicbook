## 왼손반주

지금까지는 주선율을 만드는 법에 집중 하였는데요?
이제 왼손 반주를 한번 만들어 보겠습니다.

```python
오른손은 60부터 72 까지 입니다.
C4부터 C5입니다.
```

```python
from music21 import *
import random

s1 = stream.Stream();
s1.clear()
s1.insert(0, metadata.Metadata())
s1.metadata.title = '32마디 작곡 '
s1.metadata.composer = '온석대학원대학교'


for music in range(32):
    random_int_no_again = random.sample(range(60,72),12) #60부터 72 사이에서 12개의 랜덤 수를 뽑아낸다.

    for x in random_int_no_again :
        note_or_rest = random.randrange(1, 10)
        complexity=((music/8)+4)
        if note_or_rest<(music/8)+4:
            n1 = note.Note(x)
            random_length = random.randrange(1, 4) * 0.25
            n1.quarterLength=random_length
        else:
            n1 = note.Rest(x)
            random_length = random.randrange(1, 4) * 0.25
            n1.quarterLength=random_length
        s1.append(n1)      
s1.show()
```

왼손은 36부터 48 까지 입니다.
C2부터 C3입니다.

```python
from music21 import *
import random

s2 = stream.Stream();
s2.clear()
s2.insert(0, metadata.Metadata())
s2.metadata.title = '32마디 작곡 '
s2.metadata.composer = '온석대학원대학교'


for music in range(32):
    random_int_no_again = random.sample(range(36,48),12) #60부터 72 사이에서 12개의 랜덤 수를 뽑아낸다.

    for x in random_int_no_again :
        note_or_rest = random.randrange(1, 10)
        complexity=((music/8)+4)
        if note_or_rest<(music/8)+4:
            n1 = note.Note(x)
            random_length = random.randrange(1, 4) * 0.25
            n1.quarterLength=random_length
        else:
            n1 = note.Rest(x)
            random_length = random.randrange(1, 4) * 0.25
            n1.quarterLength=random_length
        s2.append(n1)      
s2.show()
```
