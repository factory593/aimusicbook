## 32마디 작곡

이제 음악을 좀 길게 작곡 해 보겠습니다.
32마디에 도전해 봅니다.
12음렬 주의 음악은 12음을 한번씩만 사용해야 합니다.
예를들어 
C를 사용하였다면 
나머지 11개의 음을 모두 사용해야
다시 C를 사용할 수 있습니다.

그러므로 

지금만든 음악을 여러번 반복해서 32마디가 나올때 까지
프로그램을 실행 하면 되겠네요


```python
from music21 import *
import random

s1 = stream.Stream();
s1.clear()
s1.insert(0, metadata.Metadata())
s1.metadata.title = '32마디 작곡 '
s1.metadata.composer = '온석대학원대학교'


for music in range(32):
    random_int_no_again = random.sample(range(60,72),12) #60부터 72 사이에서 12개의 랜덤 수를 뽑아낸다.

    for x in random_int_no_again :
        note_or_rest = random.randrange(1, 10)
        if note_or_rest<7:
            n1 = note.Note(x)
            random_length = random.randrange(1, 4) * 0.25
            n1.quarterLength=random_length
        else:
            n1 = note.Rest(x)
            random_length = random.randrange(1, 4) * 0.25
            n1.quarterLength=random_length
        s1.append(n1)      
s1.show()
```

```python
s1.show('midi')
```
