# 코사인 함수를 사용한 왼손 


```python
from music21 import *

note_number=[]
for n in range(32):
        note_number.append((cos(n)*12)+36)

s2 = stream.Stream();
s2.clear()
s2.insert(0, metadata.Metadata())
s2.metadata.title = '코사인함수 '
s2.metadata.composer = '온석대학원대학교'

for x in note_number :
    n2 = note.Note(x)
    n2.quarterLength=1
    s2.append(n2)      
s2.show()
```

```python
s2.show("midi")
```

# 사인 코사인 에튀드

이제 음악을 조금더 길게 32마디를 만들어 보겠습니다.
32마디를 4분음표로 채우려면 
한마디에 4분음표가 4개씩이므로
4*32 = 128개의 음들이 필요하게 됩니다.

```python
from music21 import *
s1.clear()
s2.clear()

note_number1=[]
for n in range(128):
    note_number1.append((sin(n)*12)+60)
for x in note_number1 :
    n1 = note.Note(x)
    n1.quarterLength=1
    s1.append(n1)

note_number2=[]
for n in range(128):
    note_number2.append((cos(n)*12)+36)
        
for x in note_number2 :
    n2 = note.Note(x)
    n2.quarterLength=1
    s2.append(n2)  
    
twopart = stream.Stream();
twopart.clear()
twopart.insert(0, metadata.Metadata())
twopart.metadata.title = '사인 코사인 에튀드 '
twopart.metadata.composer = '온석대학원대학교'
twopart.insert(0,s1)
twopart.insert(0,s2)
twopart.show()
```

```python
twopart.show('midi')
```
