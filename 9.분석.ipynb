BTS의 음악은 왜 히트를 하는것일까요?


다음은 BTS의 다이너마이트곡의 멜로디 미디 파일 입니다


```python
from music21 import *
file = converter.parse('BTS_Dynamite_Melody.mid')
components = []
for element in file.recurse():
    components.append(element)
```

'BTS_Dynamite_Melody.mid' 파일을 convert.parser를 통하여 가져옵니다.
components = [] 를 통해 components 라는 리스트를 만들고 
file의 끝까지 iterate,반복 시키면서
componets에 element를 append(추가)합니다.

그러면 BTS의 Dynamite 멜로디가 파이썬에서 요리하기 쉽게 불러 들여 집니다.

```python
components
```

componets에는 이런데이터가 있습니다.


```python
type(components)
```

componets 의 타입은 list입니다
이제 이 components를 음악 stream 에 넣고 악보를 출력해 보겠습니다

```python
s=stream.Stream(components)
```

```python
s.show()
```

s 는 BTS Dynamite 멜로디 stream을 의미합니다.

```python
type(s)
```

s의 타입은 music21.stream.base.Stream 입니다만 기본적으로 음들의 list 입니다.
그러므로 이 s의 pitch 히스토그램을 
p = graph.plot.HistogramPitchSpace(s)
명령을 사용하여 
그려볼수 있습니다

```python
p = graph.plot.HistogramPitchSpace(s)
p.id
p.run()
```
