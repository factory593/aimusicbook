## 음의 길이 에 관하여 

음의 길이가 전부 4분 음표라서 매우 지루 한 음악이 만들어 졌습니다.
음악은 다양한 음의 길이가 존재하며 긴장과 이완을 만들어 내는데요.
이제 다양한 음의 길이를 만들어 보겠습니다.

4/4박자에서

온음표는 quarterLength=4
2분음표는 quarterLength=2
4분음표는 quarterLength=1
8분음표는 quarterLength=0.5
16분음표는 quarterLength=0.25

입니다.

지금부터 작곡할 음악의 음의 길이는 0.25의 배수를 만들겁니다.
0.25 x 1 =0.25 16분음표
0.25 x 2 =0.5   8분음표
0.25 x 3 =0.75  점8분음표
0.25 x 4 =1     4분음표

음의 길이를 정하는 난수는 1,2,3,4 가 필요합니다.
그러기 위해 다음과 같은 코드를 넣습니다.

```python
random.randrange(1, 4)
```

0부터 4까지의 수가 발생합니다 (컴퓨터에서 수는 정수, 소수등 여러가지가 있습니다만. 당분간 수는 그냥 하나의 수로 퉁 칩시다)

## 노트의 길이가 변하는 12음주의 음악 

매번 다르게 만들 수있는 12음주의 음악 2 의 코드를 정리해 보겠습니다.

```python
from music21 import *
import random

random_int_no_again = random.sample(range(60,72),12) #60부터 72 사이에서 12개의 랜덤 수를 뽑아낸다.

s1.clear()
s1 = stream.Stream();
s1.insert(0, metadata.Metadata())
s1.metadata.title = '매번 다르게 만들 수있는 12음주의 음악 3 '
s1.metadata.composer = '파이썬'

for x in random_int_no_again :
    n1 = note.Note(x)
    random_length = random.randrange(1, 4) * 0.25
    n1.quarterLength=random_length
    s1.append(n1)   
```
